```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv')
```

```python
df.columns
```

```python
sample_size=500
sample=df['alcohol'].sample(sample_size)
pop=df['alcohol']

sample_mean=sample.mean()
pop_mean=pop.mean()
error=pop_mean-sample_mean
```

```python
plt.scatter(range(1,sample_size+1),sample)
plt.axhline(pop_mean,color='red',label='Population Mean')
plt.axhline(sample_mean,color='blue',label='Sample Mean')
plt.legend()
plt.show()
```

```python
#STRATIFIED SAMPLING

#calc mean for red wine and white wine
type_group=df.groupby('type')
red=type_group.get_group('red')
white=type_group.get_group('white')

#samples of red and white
sample_size=50
red_sample=red['alcohol'].sample(sample_size)
white_sample=white['alcohol'].sample(sample_size)

#mean
pop_mean=pop_mean-red_sample
print(pop_mean)
```

```python
housing = pd.read_csv('/content/housing.csv')
```

```python
sample_size=1000
sample=housing['total_rooms'].sample(sample_size)
pop=housing['total_rooms']

sample_mean=sample.mean()
pop_mean=pop.mean()
error=pop_mean-sample_mean
```

```python
plt.scatter(range(1,sample_size+1),sample)
plt.axhline(pop_mean,color='red',label='Population Mean')
plt.axhline(sample_mean,color='blue',label='Sample Mean')
plt.legend()
plt.show()
```

```python
proximity=housing.groupby('ocean_proximity')
near_bay=proximity.get_group('NEAR BAY')
inland=proximity.get_group('INLAND')
sample_size = 50
near_bay_sample=near_bay['total_rooms'].sample(sample_size)
inland_sample=inland['total_rooms'].sample(sample_size)
near_bay_sample_mean = near_bay_sample.mean()
inland_sample_mean = inland_sample.mean()

pop_mean_near_bay=near_bay['total_rooms'].mean()
pop_mean_inland=inland['total_rooms'].mean()
```

```python
plt.scatter(range(1,sample_size+1),near_bay_sample)
plt.scatter(range(1,sample_size+1),inland_sample)
plt.axhline(pop_mean_near_bay,color='red',label='Near bay Population Mean')

plt.axhline(pop_mean_inland,color='blue',label=' Inland Population Mean')

plt.legend()
plt.show()
```
