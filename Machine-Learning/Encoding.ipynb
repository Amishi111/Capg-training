```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn import datasets
dir(datasets)
```

```python
wine=datasets.load_wine()
wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_df['target'] = wine.target
wine_df.head()
```

```python
gender_data = {'gender': ['male', 'female', 'male', 'female', 'male', 'female']}
new_df = pd.DataFrame(gender_data)
new_df
```

```python
label_encoder = LabelEncoder()
new_df['gender_encoded'] = label_encoder.fit_transform(new_df['gender'])
display(new_df)
```

```python
one = OneHotEncoder(sparse_output=False)
new_df=pd.DataFrame({'Product':['laptop','tab','watch','tab','laptop','watch']})
data=one.fit_transform(new_df)
name=one.get_feature_names_out(['Product'])
display(data)
display(name)
```

```python
ord=OrdinalEncoder(categories=[['Bad','Average','Good']])
new_df=pd.DataFrame({'Reviews':['Bad','Average','Good','Average','Good','Bad']})
new_df['Encodings']=ord.fit_transform(new_df[['Reviews']])
display(new_df)
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer()
df=pd.DataFrame({
    "height": [162,171,155,165],
    "weight": [65,60,50,62],
    "speed": [5,3,7,2]
})
data=norm.fit_transform(df)
df=pd.DataFrame(data,columns=df.columns)
display(df)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df=pd.DataFrame({
    "height": [162,171,155,165],
    "weight": [65,60,50,62],
    "speed": [5,3,7,2]
})
data=sc.fit_transform(df)
df=pd.DataFrame(data,columns=df.columns)
display(df)
```
