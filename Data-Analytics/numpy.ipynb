```python
def find_sum(arr,k):
  left=0
  right=len(arr)-1
  total=0
  while left<right:
    total = arr[left]+arr[right]
    if total == k:
      return arr[left],arr[right]
    elif total<k:
      left+=1
    else:
      right-=1
  return -1
arr=[10,20,35,50,75,80,85]
find_sum(arr,70)
```

```python
def max_sum_three_consecutive(arr):
    max_sum = float('-inf')
    current_sum = sum(arr[0:3])
    max_sum = current_sum
    result_indices = (0, 1, 2)

    for i in range(1, len(arr) - 2):
        current_sum = current_sum - arr[i-1] + arr[i+2]
        if current_sum > max_sum:
            max_sum = current_sum
            result_indices = (i, i+1, i+2)

    return max_sum, [arr[result_indices[0]], arr[result_indices[1]], arr[result_indices[2]]]

my_array = [5,2,-1,6,-2,7,3]
max_s, numbers = max_sum_three_consecutive(my_array)
print(f"The maximum sum of three consecutive numbers is: {max_s}")
print(f"The numbers are: {numbers}")
```

```python
arr=[5,2,-1,6,-2,7,3]
def window(arr,k):
  left=0
  right=0
  for right in range(len(arr)):
    total+=arr[right]
    while total>k:
      total-=arr[left]
      left+=1
    if total==k:
      return True
  return False
```
