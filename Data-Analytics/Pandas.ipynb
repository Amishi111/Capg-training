```python
#boolean filtering
#condition output is true or false || gives for all rows
#titanic['Age']>18
titanic[titanic['Age']==18]  #this shows only filtered rows
females=titanic[titanic['Sex']=='female']
females   #assigning and
```

```python
titanic[(titanic['Sex']=='female')&(titanic['Age']>18)]
titanic[((titanic['Sex']=='female')&(titanic['Age']>18))|(titanic['Fare']>500)]
```

```python
#grouping
titanic['Sex'].unique()
#grouping is done on categorical columns

titanic.groupby('Sex') #shows address only
titanic.groupby('Sex').get_group('female') #returns dataframe

#Syntax - DataFrame.groupby('ColumnName')
#use get_group to perform the grouping perfectly

titanic['Embarked'].unique()
titanic.groupby('Embarked').get_group('S')
```

```python
#aggregate functions
fare_more_than_400=titanic[titanic['Fare']>400]
fare_more_than_400.groupby('Sex').get_group('male')

fare_more_than_400['Fare'].sum()   #need to mention the specific column too
fare_more_than_400['Name'].str.upper()

titanic['Fare'].max()
titanic['Fare'].min()

titanic.groupby('Survived').get_group(1)
female_died = titanic.groupby('Survived').get_group(0).groupby('Sex').get_group('female')
female_died[female_died['Age']<18]
titanic.groupby(['Sex','Survived']).get_group(('female',0)).count()
```

```python
#drop rows or columns
#sort data
#reset index
#applying in-built function
#handling missing values
#we can drop the entire row where null/nan is there
titanic.dropna(inplace=True)
#fill nan for integer or float data type - 0,mean,mode
#fill with backward fill and forward fill
titanic['Cabin'].fillna('unknown')
titanic['Cabin'].fillna(method='ffill')
titanic['Cabin'].fillna(method='bfill')

titanic['Age'].fillna(titanic['Age'].fillna(0))
#duplicates
titanic.duplicated().value_counts()
titanic.drop_duplicates(inplace=True)
#conversion of a column's data type into another data type
titanic['Age'].astype('int')
```
